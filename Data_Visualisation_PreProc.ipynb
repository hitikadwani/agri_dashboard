## Merge Fertilizer and Area Data

```python
production_df = pd.read_csv('/content/state-level-prod.csv')
fertilizer_df = pd.read_csv('/content/state-level-fert.csv')
```

```python
def load_and_merge_datasets(production_data, fertilizer_data):

    if isinstance(production_data, str):
        prod_df = pd.read_csv(production_data, sep='\t')
    else:
        prod_df = production_data.copy()

    if isinstance(fertilizer_data, str):
        fert_df = pd.read_csv(fertilizer_data, sep='\t')
    else:
        fert_df = fertilizer_data.copy()

    prod_df.columns = [col.strip() for col in prod_df.columns]
    fert_df.columns = [col.strip() for col in fert_df.columns]

    merged_df = pd.merge(
        prod_df,
        fert_df,
        how='inner',
        on=['State Code', 'State Name', 'Year']
    )

    # Sort the data
    merged_df = merged_df.sort_values(['State Name', 'Year'])

    # Calculate fertilizer use efficiency metrics
    # Production per ton of fertilizer used
    for crop in ['Rice', 'Wheat', 'Maize', 'Chickpea', 'Groundnut']:
        prod_col = f'Avg {crop} Production (1000 tons)'
        if prod_col in merged_df.columns:
            merged_df[f'{crop} Production per Ton Fertilizer'] = (
                merged_df[prod_col] * 1000 /
                merged_df['Total State Consumption (tons)']
            )

    return merged_df
```

```python
merged_data = load_and_merge_datasets(production_df, fertilizer_df)
```

## Preprocessing Irrigation Data

```python
def aggregate_districts_to_state(df):
    """
    Aggregate district-level agricultural data to state level by calculating means

    Parameters:
    df (pandas.DataFrame): Input DataFrame with district-level data

    Returns:
    pandas.DataFrame: State-level aggregated data by year
    """

    agricultural_columns = [
        'RICE IRRIGATED AREA (1000 ha)',
        'WHEAT IRRIGATED AREA (1000 ha)',
        'MAIZE IRRIGATED AREA (1000 ha)',
        'CHICKPEA IRRIGATED AREA (1000 ha)',
        'GROUNDNUT IRRIGATED AREA (1000 ha)'
    ]

    # Group by Year and State Name, then calculate mean for agricultural columns
    state_aggregated = (df.groupby(['Year', 'State Name'])[agricultural_columns]
                         .mean()
                         .round(3)
                         .reset_index())


    state_aggregated = state_aggregated.sort_values(['Year', 'State Name'])

    return state_aggregated
```

```python
df = pd.read_csv('ICRISAT-District Level Data (2).csv')
```

```python
state_averages = aggregate_districts_to_state(df)
state_averages
```

```python
df_irig = pd.read_csv('irig.csv')
df_state = pd.read_csv('state-level-fert-prod-fin.csv')
```

```python
def merge_agricultural_data(irrigated_data, state_metrics):

    df1 = pd.DataFrame(irrigated_data)
    df2 = pd.DataFrame(state_metrics)

    # Function to standardize column names
    def standardize_column_names(df):
        return df.rename(columns=lambda x: x.strip().replace(" ", "_"))

    # Standardize column names in both dataframes
    df1 = standardize_column_names(df1)
    df2 = standardize_column_names(df2)

    # Rename columns in first dataset
    column_mapping = {
        'index': 'index',
        'Year': 'Year',
        'State Name': 'State_Name',  # Handle possible original column name
        'State_Name': 'State_Name',  # Handle if already renamed
        'Rice Irrigated': 'Rice_Irrigated',
        'Wheat Irrigated': 'Wheat_Irrigated',
        'Maize Irrigated': 'Maize_Irrigated',
        'Chickpea Irrigated': 'Chickpea_Irrigated',
        'Groundnut Irrigated': 'Groundnut_Irrigated'
    }

    existing_columns = {k: v for k, v in column_mapping.items() if k in df1.columns}
    df1 = df1.rename(columns=existing_columns)

    # Verify merge columns exist
    merge_columns = ['State_Name', 'Year']
    for col in merge_columns:
        if col not in df1.columns:
            raise KeyError(f"Column '{col}' not found in first dataset. Available columns: {df1.columns.tolist()}")
        if col not in df2.columns:
            raise KeyError(f"Column '{col}' not found in second dataset. Available columns: {df2.columns.tolist()}")

    # Perform the merge
    merged_df = pd.merge(
        df1,
        df2,
        on=['State_Name', 'Year'],
        how='inner'
    )

    duplicate_cols = [col for col in merged_df.columns if col.endswith('_x') or col.endswith('_y')]
    if duplicate_cols:
        merged_df = merged_df.drop(duplicate_cols, axis=1)

    index_cols = [col for col in merged_df.columns if col.startswith('index')]
    if index_cols:
        merged_df = merged_df.drop(index_cols, axis=1)


    merged_df = merged_df.sort_values(['State_Name', 'Year']).reset_index(drop=True)

    return merged_df

merged_data = merge_agricultural_data(df_irig, df_state)
```

```python
merged_data.to_csv('state-level-fert-prod-fin-irig.csv')
```
